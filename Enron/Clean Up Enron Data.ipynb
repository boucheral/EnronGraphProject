```python
from email.parser import Parser
from email.message import EmailMessage
import glob
import os
import re
from collections import Counter
```

```python
def parseEmailFile(fileName) :
    with open(fileName, "r") as f:
        data = f.read()

    email = Parser().parsestr(data)
    return email
```

```python

# recursive method that adds email addresses to list if it's a file or calls itself again if it's a directory
def searchDirectory3(fileString) :
    names = glob.glob(fileString + '/*')
    c = Counter()
    for name in names :
        if (os.path.isdir(name)) :
            c = c + searchDirectory3(name)
        elif (os.path.isfile(name)) :
            email = parseEmailFile(name)
            l = [x.strip() for x in email.get('To','').split(',')]
            l.extend([x.strip() for x in email.get('From','').split(',')])
            l.extend([x.strip() for x in email.get('CC','').split(',')])
            c = c + Counter(l)
    print(fileString)
    return c

emailcount = searchDirectory3('Data')
```

```python
 def emailsAndNames(addresses, names, accepted) :

    if (addresses is None or names is None) :
        return ""
    
    n = [name.strip() for name in names.split(',')]
    
    s = ""
    x = 0
    for address in [x.strip() for x in addresses.split(',')] :
        if address in accepted and x < len(n) :
            s += n[x] + ':' + address + ';'
        x += 1

    return s
    

def getAddresses(email, accepted) :
    
    addresses ={} 
    if email['from'] is not None :
        addresses['from'] = [x.strip() for x in  email['from'].split(',') if x.strip() in accepted]
    else :
        addresses['from'] = []
    if email['to'] is not None :
        addresses['to'] = [x.strip() for x in  email['to'].split(',') if x.strip() in accepted]
    else :
        addresses['to'] = []
    if email['cc'] is not None :
        addresses['cc'] = [x.strip() for x in  email['cc'].split(',') if x.strip() in accepted] 
    else : 
        addresses['cc'] = []

    return addresses

def toString(email):
    s = ''
    addresses = getAddresses(email,accepted)
    s = s + ",".join(addresses['from']) + '|' + ",".join(addresses['to']) + '|' + ",".join(addresses['cc'])
    return s
  
def checkString(s) :
    l = s.split('|')
    if l[0] and l[1] :
        return True
    else :
        return False
```

```python
def searchDirectory2(fileString) :
    emails = []
    names = glob.glob(fileString + '/*')
    for name in names :
        if (os.path.isdir(name)) :
            emails.extend(searchDirectory2(name))
        elif (os.path.isfile(name)) :
            email = parseEmailFile(name)
            a = toString(email)
            if checkString(a) :
                emails.append(a)
    return emails
```

```python
start = 'Data\\'
folders = glob.glob("Data/*")
for folder in folders :
    emails = searchDirectory2(folder)
    n = folder.replace(start,"")
    with open("Emails/" + n + '.txt', 'w') as f :
        for email in emails :
            f.write(email)
            f.write('\n')
    print(folder)
```
